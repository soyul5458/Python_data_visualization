# Matplotlib
Among the various visualization packages available in Python, matplotlib is the most fundamental library. However, it can be a bit challenging for those who are new to Python. Let's learn the terms and usage step by step.

* Python libraries for data visualization
* Matplotlib is widely used for creating simple bar graphs, line graphs, and scatter plots for data visualization purposes.
* [Gallery](https://matplotlib.org/stable/gallery/index.html)

image.png

**Terminology Explanation**
* **figure** : The space where the graph is drawn (like a piece of paper).
* **axes** : Refers to the sub-section of the Figure where the graph is drawn. In other words, it is the part within the figure where I am currently drawing the graph.
    * It controls values like Title, X-Label, Y-Label, etc. One figure can have multiple axes.
* **axis** : Refers to the sub-components of the axes. They represent the numerical lines of the graph (x-axis, y-axis).

## Two ways of drawing a graph:
1. Stateless approach: Creating figures and axes first, then accessing each axes to draw the graph.
2. Stateful approach: Automatically accessing the current figure and axes to draw the graph (the approach used in matplotlib.pyplot).


```python
# Importing the Matplotlib Library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Method 1: Using the Object-Oriented API
This method involves directly utilizing the object-oriented library implemented in Matplotlib. It is used when we want diverse control over the components of the graph.  
Create figure and axes directly. Functions are used to create graphs targeting the selected axes.

```python
# Generating arbitrary x, y data
x = np.linspace(0,1,50)
y = np.cos(4*np.pi*x)

fig = plt.figure() # Creating the figure object
ax = fig.subplots() # Creating the axes
ax.plot(x, y) # Displaying the graph
```

### Method 2: Using the Pyplot API
This method involves importing functions from the module and implementing them. It is convenient to use.  
Functions are used to create graphs targeting the current axes.

```python
x = np.linspace(0,1,50)
y = np.cos(4*np.pi*x)
plt.plot(x, y )
```

### Method 3: Combination of Object-Oriented API and Pyplot API
`plt.subplots()` returns both Figure and Axes objects simultaneously.

* Figure 1: Drawing two plots overlaid in a single axes.

```python
x = np.linspace(0,1,50)
y1 = np.cos(4*np.pi*x)
y2 = np.cos(4*np.pi*x)*np.exp(-2*x)

fig,ax = plt.subplots()
ax.plot(x,y1,'r',lw=1)
ax.plot(x,y2,'b--',lw=1)
```

* Figure 2: Drawing two plots in two separate axes.

```python
fig,(ax1,ax2) = plt.subplots(2,1)
ax1.plot(x,y1,'r',lw=1)
ax2.plot(x,y2,'b--',lw=1)
```

## Axis Range

* `plt.xlim(xmin, xmax)`, `plt.ylim(ymin, ymax)`
* `plt.axis([xmin, xmax, ymin, ymax])`
* `plt.axis(option="auto")` : Specify options
    * `"auto"` : Automatic setting
    * `"normal"` : Automatic setting
    * `"on"`: Display axes and labels
    * `"off"`: Display axes and labels
    * `"equal"` : Set the ranges and scales of each axis equally
    * `"scaled"` : Set the scale of the axes to match the graph dimensions

    * `"image"` : Use the scaled option for the axis range with respect to the data range
    * `"square"` : Set the x-axis and y-axis sizes to be equal

```python
# generating data
year = [x for x in range(2010,2024,2)]
n_std = [40, 60,55, 75, 62, 46,80]
```


## Histogram  
A histogram is a graph that represents a frequency distribution table. The horizontal axis represents intervals, and the vertical axis represents frequencies.

The bins parameter of the `hist()` function specifies the number of intervals on the horizontal axis of the histogram. The frequency and shape of the histogram change depending on the bins size.




```python
# help: ?
?plt.hist
```

```python
# Loading 'titanic' dataset from seaborn
import seaborn as sns
titanic = sns.load_dataset("titanic")

titanic.head()
```

```python
plt.hist(titanic.age);
```

```python
# Adjusting the number of bins to 15
# Setting transparency, color, and edgecolor
plt.hist(titanic.age, bins=15, alpha=0.5, color="steelblue", edgecolor="navy", histtype="bar");
```

```python
# Draw cumulative histogram
plt.hist(titanic.age, cumulative=True, label='cumulative=True')
plt.hist(titanic.age, cumulative=False, label='cumulative=False')
plt.legend(loc='upper left');
```

```python
# Specifying the type of histogram
# histtype: You can choose from 'bar', 'barstacked', 'step', 'stepfilled'; the default is 'bar'.


age_10=titanic.age-10

# 1. bar: Traditional bar-form histogram.
plt.hist((titanic.age,age_10), histtype ='bar')
plt.title('histtype - bar')
plt.figure()

# 2. barstacked: Stacked bar-form histogram with multiple datasets.
plt.hist((titanic.age,age_10), histtype ='barstacked')
plt.title('histtype - barstacked')
plt.figure()

# 3. step: Line plot with empty interiors.
plt.hist((titanic.age,age_10), histtype ='step')
plt.title('histtype - step')
plt.figure()

# 4. stepfilled: Line plot with filled interiors.
plt.hist((titanic.age,age_10), histtype ='stepfilled')
plt.title('histtype - stepfilled')
plt.figure();
```

```python
# Overlapping
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(-5, 2, 1000)
x3 = np.random.normal(5, 3, 1000)
arg = {"histtype":"stepfilled", "alpha":0.3}

plt.hist(x1, **arg)
plt.hist(x2, **arg)
plt.hist(x3, **arg);
```

## Pie Chart
Pie charts are used to visualize the overall proportion of categorical data. It provides an easy way to view the size and relative proportions of each section.

```python
titanic.head()
```

```python
x = titanic.loc[titanic["embarked"]=='S', "sex"].value_counts()
x
```

```python
plt.pie(x);
```

```python
# Adding labels.
plt.pie(x, labels=x.index);
```

```python
# Rounding the ratio to two decimal places and displaying as a percentage.
plt.pie(x, labels=x.index, autopct="%.2f%%");
```

```python
# Change the starting position and draw in a clockwise direction.
# autopct specifies the format of numbers displayed within each wedge.
# startangle sets the starting angle for drawing the wedges.
# Setting counterclock=False displays the wedges in clockwise order.
plt.pie(x, labels=x.index, autopct="%.2f%%", startangle=90, counterclock=False);
```

```python
# Detaching and emphasizing slices.
# explode sets how far each wedge is detached from the center of the pie chart.
plt.pie(x, labels=x.index, explode=[0.1, 0.1]);
```

```python
# Adding a legend.
plt.pie(x, labels=x.index)
plt.legend(loc=(1.1, 0.5), title="Sex");
```

```python
# Specifying colors.
colors = ['silver','gold']
plt.pie(x, labels=x.index, colors=colors);
```
